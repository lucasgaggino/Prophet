```python
import datetime
import math
import random
import time


def generate_traffic_data(duration_hours=24, interval_minutes=5, traffic_type='medium',single_point=False):
    """
    Generates simulated network traffic data over a specified duration and interval.

    Args:
        duration_hours (int): The total duration in hours for which to generate data.
        interval_minutes (int): The interval in minutes between each data point.
        traffic_type (str): The type of traffic pattern to simulate ('low', 'medium', 'high', 'very high').

    Returns:
        list of dict: A list containing dictionaries with 'time' and 'traffic_value' keys.
    """
    # Get current time
    now = datetime.datetime.now() + datetime.timedelta(hours=3)

    # Calculate total number of data points
    total_minutes = duration_hours * 60
    num_points = int(total_minutes / interval_minutes) + 1

    # Generate time points
    time_points = [
        now - datetime.timedelta(minutes=interval_minutes * i)
        for i in reversed(range(num_points))
    ]

    traffic_data = []

    for time_point in time_points:
        # Convert time to fractional hours
        t = (
            time_point.hour
            + time_point.minute / 60.0
            + time_point.second / 3600.0
        )

        # Map time to sine function to simulate daily traffic patterns
        angle_o = (t / 24.0) * 2 * math.pi - math.pi / 2 
        angle = angle_o*random.uniform(0.9, 1.05)

        base_value = (math.sin(angle) + math.sin(angle*2) / 2 + math.sin(angle*3) / 3 + 1)

        # Adjust amplitude based on traffic_type
        amplitude_dict = {
            'low': 1,
            'medium': 2,
            'high': 3,
            'very high': 4
        }
        amplitude = amplitude_dict.get(traffic_type.lower(), 2)  # Default to 'medium'

        # Calculate traffic value
        traffic_value = (base_value + 1) / 2  # Normalize to 0..1
        traffic_value *= amplitude  # Scale by amplitude

        # Add random noise
        noise = random.uniform(-0.01, 0.02) * amplitude
        traffic_value += noise

        # Ensure traffic_value is non-negative
        traffic_value = max(0, traffic_value)

        # Append to traffic data
        traffic_data.append({
            'time': time_point,
            'traffic_value': traffic_value
        })

    if single_point:
        return traffic_data[-1]['traffic_value']
    return traffic_data
```
