```python
import pandas as pd 
import numpy as np
import json
from datetime import datetime
import matplotlib.pyplot as plt

```

```python
def shift_frequency(df, factor,time_column='ds'):
    """
    Shift the main frequency of a DataFrame by a given factor.

    :param df: DataFrame with 'y' and 'ds' columns, where 'ds' is the timestamp column.
    :param factor: The factor by which to shift the frequency (e.g., 2 for doubling the interval).
    :return: DataFrame with shifted frequency.
    """
    # Ensure the 'ds' column is in datetime format
    df[time_column] = pd.to_datetime(df[time_column])
    
    # Calculate the time delta between consecutive 'ds' values
    time_deltas = df[time_column].diff().dropna()
    
    # Get the mean time delta
    mean_delta = time_deltas.mean()
    
    # Apply the frequency shift
    df[time_column] = df[time_column].iloc[0] + (df.index * mean_delta * factor)
    
    return df
```

```python
data = json.load(open('data.json'))
data = data['data']['Metrics_trafico_por_sector_']
df = pd.DataFrame(data)[:-240]
df['ds']  = pd.to_datetime(df['time'])
df['ds'] = df['ds'].dt.tz_localize(None)
df.drop(columns=['time','cell'], inplace=True)
df.rename(columns={'trafficDUserDlGb': 'y'}, inplace=True)
df.dropna(inplace=True)
df['y'] = df['y'].rolling(window=3, center=True).mean()

df = shift_frequency(df, 1)
df.dropna(inplace=True)

df.to_csv('real_data.csv', index=False)
df
```
